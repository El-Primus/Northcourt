```python
# Importing Packages
import os
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
import re
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Image
```

```python
# Scraping the website
article_1 = requests.get("https://www.knightfrank.com/research/article/2022-06-29-cote-divoire")
```

```python
soup1 = bs(article_1.content, 'lxml')
```

```python
#Getting the Images
images = soup1.select('img')
images = images[1:4]
images
```

### Downloading the images

```python
url = "https://www.knightfrank.com/research/article/2022-06-29-cote-divoire"

# Creating a new directory to store the images (if it doesn't already exist)
if not os.path.exists("Images"):
    os.mkdir("Images")

# Looping through all the image tags and download the images
for image in images:
    image_url = image["src"]
    full_url = url + image_url
    filename = image_url.split("/")[-1]
    response = requests.get(full_url)
    with open(f"images/{filename}", "wb") as f:
        f.write(response.content)
```

```python
# Setting the directory where the images are saved
directory = 'Images'

# Looping through all files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.jpg') or filename.endswith('.png'):
        # Open the image
        image = Image(os.path.join(directory, filename))
        # Display the image
        display(image)
        #image.show()
```

From the images, it is safe to conclude that the most common property use types in Cote D'ivoire are
* Residential
* Industrial
* Office Spaces
* Retail

They are also the most profitable uses too
